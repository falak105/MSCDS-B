```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load data
df = pd.read_csv('CAC2_Assignment_ResponseSheet.csv')
df
```

```python
# Value counts for categorical vars
df['Stay Location'].value_counts()
df['Which source you depend more for the internet?'].value_counts()
# Bar plot for internet source
source_counts = df['Which source you depend more for the internet?'].value_counts()
ax = source_counts.plot.pie()
```

```python
# Calculate value counts for the 'Where you able to access all required educational content using College Wi-Fi?' 
edu_counts = df['Where you able to access all required educational content using College Wi-F?'].value_counts()
print(edu_counts)
# Create a bar plot to visualize the distribution
ax = edu_counts.plot.bar(rot=0)
ax.set_title('Educational Content Access')
```

```python


# Rate your experience on college Wi-Fi
value_counts = df['Rate your experience on college Wi-Fi'].value_counts()

# Sort the counts by the index (rating values)
experience_counts = value_counts.sort_index()

# Create a line chart
plt.figure(figsize=(10, 6))
experience_counts.plot(kind='line', marker='o', color='lightgreen')
plt.title('Experience with College Wi-Fi')
plt.xlabel('Rating')
plt.ylabel('Number of Respondents')
plt.xticks(experience_counts.index, rotation=0)
plt.grid(True)
plt.show()
```

```python


# Count the occurrences of each value in the specified column
value_counts = df['Over the past 1 month, how many times have you experienced complete outage of mobile network?'].value_counts()

# Sort the counts by the index (number of outages)
value_counts = value_counts.sort_index()

# Set the Seaborn style to 'whitegrid'
sns.set_style("whitegrid")

# Create a line graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=value_counts.index, y=value_counts.values, color='orange', marker='o', markersize=8)
plt.title('Mobile Network Outage in the Last Month')
plt.xlabel('Number of Outages')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=0)
plt.show()
```

```python

# Create a pivot table to count occurrences based on 'Gender' and 'Location'
heatmap_data = df.pivot_table(index='Gender', columns='Stay Location', values='Have network issues ever affected your assignment submission or your preparation for exams?', aggfunc='count')

plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')

# Create a heatmap
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu')

plt.title('Impact of Network Issues by Gender and Location')
plt.xlabel('Stay Location')
plt.ylabel('Gender')

plt.show()
```

```python
#count of genders using the location
plt.figure(figsize=(10, 6))
sns.set_style('whitegrid')

# Create a boxplot
sns.boxplot(data=df, x='Gender', y='Stay Location', palette='Set2')

plt.xlabel('Gender')
plt.ylabel('NumericVariable')
plt.legend(title='Location', loc='upper right')

plt.show()
```
